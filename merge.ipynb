```python
import pandas as pd
import glob
import os


folders = {
    "enrolment": "api_data_aadhar_enrolment",
    "demographic": "api_data_aadhar_demographic",
    "biometric": "api_data_aadhar_biometric"
}
```

```python
def merge_files_in_folder(folder_path, category_name):
    file_paths = glob.glob(os.path.join(folder_path, "*.csv"))

    dfs = []
    first_cols = None
    
    for file in file_paths:
        df = pd.read_csv(file)
        
        if first_cols is None:
            first_cols = list(df.columns)
        else:
            if list(df.columns) != first_cols:
                print(f"Error: Columns do not match in file {file}")
                continue
        
        dfs.append(df)
    
    merged_df = pd.concat(dfs, ignore_index=True)
    print(f"Successfully merged {category_name}. Total rows: {len(merged_df)}")
    return merged_df
```

```python

df_enrolment = merge_files_in_folder(folders["enrolment"], "Enrolment")


df_demographic = merge_files_in_folder(folders["demographic"], "Demographic")


df_biometric = merge_files_in_folder(folders["biometric"], "Biometric")
```

```python
print("Enrolment")
print(df_enrolment.count())

print("\nDemographic")
print(df_demographic.count())

print("\nBiometric")
print(df_biometric.count())
```

```python
df_enrolment.to_csv('enrolment.csv',index=False)
df_demographic.to_csv('demographic.csv',index=False)
df_biometric.to_csv('biometric.csv',index=False)
```
